# Funtion "format_k_no_k" that allow to eliminate the siffix "k" from the numeric data

```python
def format_k_no_k(list_2,listt):
    for i in range(0,len(listt)):
        if listt[i].find('K') == -1:        
            list_2.append(int(listt[i]))            
        else:       
            if listt[i].find('.') == -1:
                list_2.append(int(listt[i][:-1] + '000'))                        
            else:
                tmp = listt[i].replace('.','')
                list_2.append(int(tmp[:-1] + '00'))
```

# Use "format_k_no_k" in the categories "Number of Reviews", "Plays", "Playing","Backlogs"and "Wishlist"

```python
format_number_of_review = list()
format_k_no_k(format_number_of_review,df['Number of Reviews'])
format_plays = list()
format_k_no_k(format_plays,df['Plays'])
format_playing = list()
format_k_no_k(format_playing,df['Playing'])
format_backlogs = list()
format_k_no_k(format_backlogs,df['Backlogs'])
format_wishlist = list()
format_k_no_k(format_wishlist,df['Wishlist'])
```

```python
df['Number of Reviews'] = format_number_of_review
df['Plays'] = format_plays
df['Playing'] = format_playing
df['Backlogs'] = format_backlogs
df['Wishlist'] = format_wishlist
```

# The funtion "format_split" divide a column in three columns for the coma "," use the funtion in column "Gender" and "Team"

```python
one = list()
two = list()
three = list()

def format_split(df_,one_,two_,three_):
    for i in range(0,len(df)):
        s = df_[i]
        if s.count(',') == 0:
            f = s.find(',')
            one_.append(s[2:f-1])
            two_.append(0)
            three.append(0)
        elif s.count(',') == 1:
            f = s.find(',')
            se = s.find(',',f+2)
            one_.append(s[2:f-1])
            two_.append(s[f+3:se-1])
            three.append(0)
        else:
            f = s.find(',')
            se = s.find(',',f+2)
            t = s.find(',',se+1)
            one_.append(s[2:f-1])
            two_.append(s[f+3:se-1])
            three_.append(s[se+3:t-1])
            
format_split(df['Genres'],one,two,three)
df['Gender_1'] = one
df['Gender_2'] = two
df['Gender_3'] = three
df.drop(['Genres'],axis=1,inplace = True)

one.clear()
two.clear()
three.clear()

format_split(df['Team'],one,two,three)
df['Team_1'] = one
df['Team_2'] = two
df['Team_3'] = three
df.drop(['Team'],axis=1,inplace = True)
```
